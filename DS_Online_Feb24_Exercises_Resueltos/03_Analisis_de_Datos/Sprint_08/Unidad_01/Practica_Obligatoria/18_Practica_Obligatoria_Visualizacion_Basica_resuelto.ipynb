```python
# Seleccionar las 5 ciudades más pobladas
top_cities = data.nlargest(5, 'population_total')

# Dividir los valores de población por 1000
top_cities['population_total'] = top_cities['population_total'] / 1000

# Definir colores
colors = ['blue', 'gray', 'olive', 'brown', 'teal']

# Crear el gráfico de barras
plt.figure(figsize=(7, 7))
sns.barplot(x='city', y='population_total', data=top_cities, hue='city', palette=colors, alpha=0.7, legend=False)
plt.xlabel('Ciudades')
plt.ylabel('Habitantes (miles)')
plt.title('Las 5 ciudades más pobladas')

# Añadir etiquetas de población centradas
for i, population in enumerate(top_cities['population_total']):
    plt.text(i, population, str(round(population, 3)), ha='center')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convertir el área total ocupada de millas cuadradas a kilómetros cuadrados
data['area_total_km2'] = data['area_total_sq_mi'] * 2.58999

# Crear el gráfico de histograma sin bordes en las barras
plt.figure(figsize=(8, 6))
sns.histplot(data['area_total_km2'], bins=50, color='steelblue', edgecolor='none')
plt.xlabel('Superficie (Km²)')
plt.ylabel('Número de ciudades')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Establecer Comic Sans MS como fuente y tamaño de letra
plt.rcParams['font.family'] = 'Comic Sans MS'
plt.rcParams['font.size'] = 12

# Agregar leyenda
plt.legend(['Hist Area'],fontsize=8)

plt.show()
```

```python
# 3. El scatter de área de agua de un ciudad con su poblacion.

# Convertir la población a miles de habitantes
data['population_total'] = data['population_total'] / 1000

# Crear el scatter plot de área de agua vs población
plt.figure(figsize=(8, 8))
sns.scatterplot(x='area_water_km2', y='population_total', data=data, color='green')
plt.xlabel('Área de agua en la ciudad (km²)', fontsize=8)
plt.ylabel('Población (miles de habitantes)', fontsize=8)  # Reducir el tamaño de la letra
plt.title('Población por kilómetro cuadrado de agua', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)  # Reducir el tamaño de la letra
plt.gca().tick_params(axis='y', labelrotation=45)
plt.tight_layout()

# Establecer Arial como fuente
plt.rcParams['font.family'] = 'Arial'

# Encontrar los puntos outliers
outliers = data.nlargest(2, 'population_total')

# Agregar anotación para los outliers
plt.annotate('Outliners', xy=(outliers['area_water_km2'].mean(), outliers['population_total'].mean()),
             xytext=(-50, 20), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='black'))

# Agregar flechas a cada punto outlier
for i, row in outliers.iterrows():
    plt.annotate('', xy=(row['area_water_km2'], row['population_total']),
                 xytext=(outliers['area_water_km2'].mean(), outliers['population_total'].mean()),
                 arrowprops=dict(arrowstyle='->', color='red'))

plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar las 5 ciudades más pobladas
top_cities = data.nlargest(5, 'population_total')
top_cities['population_total'] = top_cities['population_total'] / 1000
colors = ['blue', 'gray', 'olive', 'brown', 'teal']

# Convertir el área total ocupada de millas cuadradas a kilómetros cuadrados
data['area_total_km2'] = data['area_total_sq_mi'] * 2.58999

# Convertir la población a miles de habitantes
data['population_total'] = data['population_total'] / 1000

# Crear el scatter plot de área de agua vs población
plt.figure(figsize=(12, 12))

# Primer gráfico: Las 5 ciudades más pobladas
plt.subplot(2, 2, 1)
sns.barplot(x='city', y='population_total', data=top_cities, hue='city', palette=colors, alpha=0.7, legend=False)
plt.xlabel('Ciudades')
plt.ylabel('Habitantes (miles)')
plt.title('Las 5 ciudades más pobladas')
for i, population in enumerate(top_cities['population_total']):
    plt.text(i, population, str(round(population, 3)), ha='center')
plt.xticks(rotation=45)

# Segundo gráfico: Histograma de área total
plt.subplot(2, 2, 2)
sns.histplot(data['area_total_km2'], bins=50, color='steelblue', edgecolor='none')
plt.xlabel('Superficie (Km²)')
plt.ylabel('Número de ciudades')
plt.title('Histograma de área total')

# Tercer gráfico: Scatter plot de área de agua vs población
plt.subplot(2, 2, 3)
sns.scatterplot(x='area_water_km2', y='population_total', data=data, color='green')
plt.xlabel('Área de agua en la ciudad (km²)', fontsize=8)
plt.ylabel('Población (miles de habitantes)', fontsize=8)
plt.title('Población por kilómetro cuadrado de agua', fontsize=12)
plt.gca().tick_params(axis='y', labelrotation=45)
outliers = data.nlargest(2, 'population_total')
plt.annotate('Outliers', xy=(outliers['area_water_km2'].mean(), outliers['population_total'].mean()),
             xytext=(-70, 20), textcoords='offset points', ha='right',
             arrowprops=dict(arrowstyle='->', color='black'))
for i, row in outliers.iterrows():
    plt.annotate('', xy=(row['area_water_km2'], row['population_total']),
                 xytext=(outliers['area_water_km2'].mean(), outliers['population_total'].mean()),
                 arrowprops=dict(arrowstyle='->', color='black'))

# Cuarto gráfico: Boxplot de altitud de las ciudades
plt.subplot(2, 2, 4)
sns.boxplot(x=data['elevation_m'], color='red', boxprops=dict(facecolor='red'))
plt.xlabel('Altitud (m)', fontsize=12)
plt.title('Diagrama de caja de la Altitud de las ciudades de California', fontsize=14)
plt.xlim(-100, 2500)

plt.tight_layout()
plt.show()
```
