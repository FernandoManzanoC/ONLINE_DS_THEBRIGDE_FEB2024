### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Selecciona las columnas "city" y "population_total" del DataFrame casas
top_five = casas[["city","population_total"]]

# Ordena el DataFrame segn la poblacin total de manera descendente
top_five = top_five.sort_values(by="population_total", ascending=False)

# Calcula la poblacin en miles y agrega como una nueva columna
top_five["Poblacion en miles"] = top_five["population_total"]/1000

# Selecciona las columnas "city" y "Poblacion en miles"
top_five[["city","Poblacion en miles"]] 

# Selecciona las cinco primeras filas del DataFrame
df_top_five = top_five.head(5)

# Extrae las ciudades y la poblacin de las cinco ciudades ms pobladas
ciudades = df_top_five["city"]
poblacion = df_top_five["Poblacion en miles"]

# Define los colores para las barras del grfico
colores = ['blue', 'grey',"red", 'olive','brown']

# Crea un grfico de barras con las ciudades ms pobladas
plt.bar(ciudades, poblacion, color=colores, alpha=0.6),

# Establece los lmites y marcas en el eje y
plt.yticks(np.arange(0, 5001, 1000)),

# Establece el ttulo del grfico
plt.title("Ciudades ms pobladas de California", fontsize=14)

# Etiqueta el eje x del grfico
plt.xlabel("Ciudades", fontsize=12),

# Etiqueta el eje y del grfico
plt.ylabel("Habitantes (miles)", fontsize=12),

# Aade etiquetas con la poblacin en miles a las barras correspondientes
for index, row in df_top_five.iterrows():
    plt.text(row["city"], row["Poblacion en miles"] + 150, str(row["Poblacion en miles"]), ha='center')


```

```python
# Selecciona los datos del rea total en kilmetros cuadrados del DataFrame casas
df_histo = casas["area_total_km2"]

# Crea un histograma con los datos del rea total
plt.hist(df_histo, bins=50, color="steelblue", label="Hist Area", alpha=0.7)

# Etiqueta el eje x del histograma
plt.xlabel("Superficie ($km^2$)", fontname='Comic Sans MS', fontsize=12)

# Etiqueta el eje y del histograma
plt.ylabel("Nmero de ciudades", fontname='Comic Sans MS', fontsize=12)

# Establece el ttulo del histograma
plt.title('rea total ocupada por las ciudades', fontname='Comic Sans MS', fontsize=12)

# Muestra la leyenda del histograma
plt.legend();


```

```python
# Selecciona las columnas relevantes del DataFrame casas
df_poblacionxkm2 = casas[["city","population_total","area_water_km2"]]

# Ordena el DataFrame segn la poblacin total de manera descendente
df_poblacionxkm2 = df_poblacionxkm2.sort_values(by="population_total", ascending=False)

# Calcula la poblacin en miles y agrega como una nueva columna
df_poblacionxkm2["Poblacion en miles"] = df_poblacionxkm2["population_total"]/1000

# Elimina las columnas que ya no son necesarias
df_poblacionxkm2= df_poblacionxkm2.drop(columns=["population_total"])
df_poblacionxkm2= df_poblacionxkm2.drop(columns=["city"])

# Selecciona las columnas relevantes para el anlisis posterior
df_poblacionxkm2[["Poblacion en miles","area_water_km2"]] 

# Encuentra el ndice del valor mximo de poblacin y de rea de agua
max_poblacion = df_poblacionxkm2["Poblacion en miles"].idxmax()
max_agua_area = df_poblacionxkm2["area_water_km2"].idxmax()

# Obtiene las coordenadas de los outliers (valores mximos)
y_outliner = df_poblacionxkm2.loc[max_poblacion, "Poblacion en miles"]
x_outliner = df_poblacionxkm2.loc[max_poblacion, "area_water_km2"]
y_outliner_max_area = df_poblacionxkm2.loc[max_agua_area, "Poblacion en miles"]
x_outliner_max_area = df_poblacionxkm2.loc[max_agua_area, "area_water_km2"]

# Crea un diagrama de dispersin con las columnas seleccionadas
plt.scatter(df_poblacionxkm2["area_water_km2"],df_poblacionxkm2["Poblacion en miles"] , alpha= 0.55, color="green")

# Aade anotaciones para marcar los outliers en el grfico
plt.annotate("Outlier", xy=(x_outliner, y_outliner),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

plt.annotate("Outlier", xy=(x_outliner_max_area, y_outliner_max_area),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

# Aade ttulo y etiquetas a los ejes del grfico
plt.title("Poblacin por kilmetro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad (km2)")
plt.ylabel("Poblacin (miles de habitantes)");
```

```python
# Selecciona los datos de altitud del DataFrame casas
data_box = casas["elevation_m"]

# Crea un diagrama de caja horizontal con los datos de altitud
plt.boxplot(data_box, vert=False, patch_artist=True, boxprops={"facecolor": "red", "edgecolor": "black"}, medianprops={"color": "black"})

# Etiqueta el eje x del grfico
plt.xlabel("Altitud (m)", fontsize=6);
plt.title('Diagrama de caja de la Altitud de las ciudades de California', fontsize=14)
```

```python
# Crea una figura y ejes con 2 filas y 2 columnas, con un tamao de 20x20 pulgadas
fig, ax = plt.subplots(2, 2, figsize=(20, 20))

# Crea un array de texto lleno de valores None con la misma forma que los ejes
textos = np.full(fill_value=None, shape=ax.shape)

# Establece el color de fondo de la figura como blanco
plt.gcf().set_facecolor('white') 

# Primer subgrfico: diagrama de barras de las ciudades ms pobladas de California
ax[0, 0].bar(ciudades, poblacion, color=colores, alpha=0.6)
ax[0, 0].set_yticks(np.arange(0, 5001, 1000))
ax[0, 0].set_title("Ciudades ms pobladas de California", fontsize=15)
ax[0, 0].set_xlabel("Ciudades", fontsize=12)
ax[0, 0].set_ylabel("Habitantes (miles)", fontsize=12)

# Aade etiquetas con la poblacin en miles a las barras correspondientes
for index, row in df_top_five.iterrows():
    ax[0, 0].text(row["city"], row["Poblacion en miles"] + 150, str(row["Poblacion en miles"]), ha='center') 

# Segundo subgrfico: histograma del rea de las ciudades
ax[0, 1].hist(df_histo, bins=50, color="steelblue", label="Hist Area", alpha=0.9)
ax[0, 1].set_xlabel("Superficie ($km^2$)", fontname='Comic Sans MS', fontsize=18)
ax[0, 1].set_ylabel("Nmero de ciudades", fontname='Comic Sans MS', fontsize=18)
ax[0,1].set_title('rea total ocupada por las ciudades', fontname='Comic Sans MS', fontsize=12)
ax[0, 1].legend() 

# Tercer subgrfico: diagrama de dispersin de poblacin vs rea de agua
ax[1,0].scatter(df_poblacionxkm2["area_water_km2"], df_poblacionxkm2["Poblacion en miles"], alpha=0.55, color="green")

# Aade anotaciones para los outliers en el grfico de dispersin
ax[1,0].annotate("Outlier", xy=(x_outliner, y_outliner),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

ax[1,0].annotate("Outlier", xy=(x_outliner_max_area, y_outliner_max_area),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

ax[1,0].set_title("Poblacin por kilmetro cuadrado de agua", fontsize=18)
ax[1,0].set_xlabel("rea de agua en la ciudad (km2)", fontsize=12)
ax[1,0].set_ylabel("Poblacin (miles de habitantes)", fontsize=12)

# Cuarto subgrfico: diagrama de caja de la altitud de las ciudades
ax[1,1].boxplot(data_box, vert=False, patch_artist=True, boxprops={"facecolor": "red", "edgecolor": "black"}, medianprops={"color": "black"})
ax[1,1].set_xlabel("Altitud (m)", fontsize=12)
ax[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California", fontsize=18)
# Ajusta el diseo para evitar solapamientos
plt.tight_layout()

# Guarda la figura como "cuatro_en_uno.png"
fig.savefig("cuatro_en_uno.png")
```
