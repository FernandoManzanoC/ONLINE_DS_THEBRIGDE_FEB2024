## #2 Texto

Vamos a leer ahora el fichero indicado en la ruta contenida en `fichero_texto`, el objetivo es procesar el contenido para construir un `DataFrame` con las siguientes columnas direcciones de correo del emisor, veredicto de si es spam o no y confidencia en ese veredicto.

1. Crea un diccionario con las siguientes claves: "from","to","is_spam","confidence" y que cada una tenga una lista vacia.

```python
# Diccionario para almacenar los datos
datos = {
    "from": [],
    "to": [],
    "is_spam": [],
    "confidence": []
}
```

2. Lee el fichero (usando `with open...`) y muestra las primeras 60 filas por pantalla, as podrs ver como es ms o menos la estructura del mismo

```python
# Ruta del archivo de texto
fichero_texto = "./data/mbox.txt"

# Leer las primeras 60 lneas del archivo
with open(fichero_texto, "r", encoding="utf-8") as archivo:
    for i, linea in enumerate(archivo):
        if i < 60:
            print(linea.rstrip())  # Mostrar la lnea sin saltos de lnea adicionales
        else:
            break  # Salir del bucle despus de mostrar las primeras 60 lneas
```

3. Ahora sabemos que el emisor viene despus de "To:" el receptor despus de "From:" el resultado de Spam tras "X-DSPAM-Result" y el nivel de confianza tras "X-DSPAM-Confidence", crea un programa que lea el fichero y a medida que va leyendo el texto aada cada valor (emisor, receptor, si es spam o no, nivel de confianza) a cada una de las listas del diccionario creado en 1.  
Por ejemplo, cuando llegue a una lnea con "To:" que aada el correo a <diccionario>["to"]

```python

# Leer el archivo de texto y procesar las lneas
with open(fichero_texto, "r", encoding="utf-8") as archivo:
    for linea in archivo:
        # Extraer el emisor
        if linea.startswith("From:"):
            emisor = linea.split(":")[1].strip()
            datos["from"].append(emisor)
        # Extraer el receptor
        elif linea.startswith("To:"):
            receptor = linea.split(":")[1].strip()
            datos["to"].append(receptor)
        # Verificar si es spam
        elif linea.startswith("X-DSPAM-Result:"):
            resultado_spam = linea.split(":")[1].strip()
            if resultado_spam == "Innocent":
                datos["is_spam"].append(False)
            else:
                datos["is_spam"].append(True)
        # Extraer el nivel de confianza
        elif linea.startswith("X-DSPAM-Confidence:"):
            confianza = float(linea.split(":")[1].strip())
            datos["confidence"].append(confianza)

# Verificar la longitud de todas las listas para asegurarnos de que tengan la misma longitud
longitudes = [len(datos["from"]), len(datos["to"]), len(datos["is_spam"]), len(datos["confidence"])]
if len(set(longitudes)) != 1:
    print("Error! Las listas no tienen la misma longitud.")
else:
    print("Datos procesados con xito!")

# Mostrar la cantidad de datos ledos
print("Cantidad de correos procesados:", longitudes[0])
```

5. Ahora ya crea un dataframe a partir del diccionario y muestra si hay algn correo declarado como spam y la confianza media. (quiz tengas que cambiar el tipo de la columna)

```python

import pandas as pd

# Convertir la lista de valores de "is_spam" a tipo booleano
datos["is_spam"] = [bool(valor) for valor in datos["is_spam"]]

# Crear DataFrame
df = pd.DataFrame(datos)

# Mostrar si hay algn correo declarado como spam y la confianza media
if df['is_spam'].any():
    confianza_media_spam = df[df['is_spam']]['confidence'].mean()
    print("Hay correos declarados como spam.")
    print("Confianza media de los correos spam:", confianza_media_spam)
else:
    print("No hay correos declarados como spam en el DataFrame.")

```

2. Asigna el nodo raz a la variable `ejer_raiz` y muestra su etiqueta (y sus atributos, si es que tiene)

```python

# Obtener el nodo raz
ejer_raiz = arbol.getroot()

# Mostrar la etiqueta del nodo raz
print("Etiqueta del nodo raz:", ejer_raiz.tag)

# Mostrar los atributos del nodo raz, si los tiene
if ejer_raiz.attrib:
    print("Atributos del nodo raz:", ejer_raiz.attrib)
else:
    print("El nodo raz no tiene atributos.")
```

3. Muestra cuantos hijos y nietos tiene el nodo raiz y la etiqueta de cada uno:

```python
# Contadores de hijos y nietos
num_hijos = len(ejer_raiz)
num_nietos = sum(len(hijo) for hijo in ejer_raiz)

# Mostrar el nmero de hijos y nietos del nodo raz
print("El nodo raz tiene", num_hijos, "hijos y", num_nietos, "nietos.")

# Mostrar la etiqueta de cada hijo y nieto
print("Etiquetas de los hijos del nodo raz:")
for hijo in ejer_raiz:
    print("-", hijo.tag)
    for nieto in hijo:
        print("--", nieto.tag)
```

4. Habrs comprobado que del nodo raz cuelgan 1673 nodos con tag "Event", encuentra uno (<nodo_raiz>.find("Event"), pero primero tendrs que encontrar un nodo raiz, pista raiz.find("Game") y de ahi...) y pinta sus atributos


```python
# Encontrar el primer nodo <Game>
nodo_game = ejer_raiz.find("Game")

# Encontrar el primer nodo <Event> dentro del nodo <Game>
nodo_event = nodo_game.find("Event")

# Mostrar los atributos del nodo <Event>
if nodo_event is not None:
    print("Atributos del nodo <Event>:")
    for atributo, valor in nodo_event.attrib.items():
        print(f"{atributo}: {valor}")
else:
    print("No se encontr ningn nodo <Event> dentro del nodo <Game>.")

```

5. Recorre de nuevo los eventos y crea un dataframe con las columnas de las claves del attrib de un "Event" y como valores los valores de los atributos de los 1673 eventos que hay colgando del nodo raz.

```python
import pandas as pd

# Crear una lista para almacenar los atributos de cada evento
atributos_eventos = []

# Iterar sobre los nodos <Event>
for evento in ejer_raiz.iter("Event"):
    # Obtener los atributos del evento y almacenarlos en un diccionario
    atributos = evento.attrib
    atributos_eventos.append(atributos)

# Crear DataFrame a partir de la lista de atributos de eventos
df_eventos = pd.DataFrame(atributos_eventos)

# Mostrar las primeras filas del DataFrame
print(df_eventos.head())
```
